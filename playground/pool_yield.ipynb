```python
from urllib.request import urlopen
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
url_pools = "https://yields.llama.fi/pools"
response = urlopen(url_pools)
data_json = json.loads(response.read())
df_data = data_json['data']
df = pd.DataFrame(df_data)
df
```

```python
df_eth = df[df['chain'] == "Ethereum"] 
df_eth
```

```python
df_eth_sorted = df_eth.sort_values('tvlUsd', ascending=False).reset_index()
df_eth_sorted
```

```python
url_list = []
for i in range(5):
    buffer = 'https://yields.llama.fi/chart/'
    url = buffer + df_eth_sorted['pool'][i]
    url_list.append(url)
print(url_list)
```

```python
response = urlopen(url_list[0])
data_json = json.loads(response.read())
df1 =  pd.DataFrame(data_json['data'])
df1['timestamp'] = pd.to_datetime(df1['timestamp'])
df1 = df1.set_index('timestamp')
df1.apy.plot()
```
